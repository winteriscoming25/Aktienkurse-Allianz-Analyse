```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = "ALV.DE"
ticker1 = "CS.PA"
ticker2 = "G.MI"
ticker3 = "CB"
```

```python
yf.download(ticker)
```

```python
yf.download(ticker1)
```

```python
yf.download(ticker2)
```

```python
yf.download(ticker3)
```

```python
ALLIANZ = yf.download(ticker, start="1996-01-01", end="2020-12-01")
AXA = yf.download(ticker1, start="1996-01-01", end="2020-12-01")
GENERALI = yf.download(ticker2, start="1996-01-01", end="2020-12-01")
CHUBB = yf.download(ticker3, start="1996-01-01", end="2020-12-01")
```

```python
y1=ALLIANZ['Close']
y2=AXA ['Close']
y3=GENERALI['Close']
y4=CHUBB['Close']
```

```python
import seaborn
y1.plot(figsize=(16,9), color ='#1155dd', label='Close_ALLIANZ')
y2.plot(figsize=(16,9), color ="#15cc55", label='Close_AXA')
y3.plot(figsize=(16,9), color = "red", label='Close_GENERALI')
y4.plot(figsize=(16,9), color = "#f7f420", label='Close_CHUBB')


plt.legend(loc='upper right')


plt.xlabel("Zeitraum (Täglich)", fontsize=14)
plt.ylabel("Aktienkurs (Euros/Dollars)", fontsize=14)
plt.title("Aktienkursentwicklung zwischen 1996 und 2020", fontsize=20)




plt.show()
```

```python
ALLIANZ_PANDEMIE= yf.download(ticker, start = "2019-10-01", end ="2020-12-11", interval="1wk")
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
ALLIANZ_PANDEMIE = yf.download(ticker, start = "2019-10-01", 
                               end ="2020-12-11", interval="1wk", auto_adjust=True, rounding = True)
```

```python
print(ALLIANZ_PANDEMIE)
```

```python
ALLIANZ_PANDEMIE['Close'].plot(figsize=(16,9), color ='blue', label="Close_ALLIANZ", marker ="o")

plt.xlabel("Zeitraum (wöchlich)", fontsize=14)
plt.ylabel("Aktienkurs (Euros)", fontsize=14)
plt.title("Aktienkursentwicklung während der Pandemie", fontsize=20)

plt.legend(loc='upper right')

plt.show()
```
